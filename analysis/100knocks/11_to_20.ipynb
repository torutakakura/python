```python
import pandas as pd

# csvファイル読み込み
DATASET_PATH = 'C://Users/toru/Work/dataset/11-20/'
target = "uriage.csv"
path = DATASET_PATH + target
uriage_data = pd.read_csv(path)
uriage_data.head()
```

```python
# excelファイルの読み込み
target = "kokyaku_daicho.xlsx"
path = DATASET_PATH + target
kokyaku_daicho_data = pd.read_excel(path)
kokyaku_daicho_data.head()
```

```python
uriage_data["purchase_date"] = pd.to_datetime(uriage_data["purchase_date"])
uriage_data["purchase_date"].head()
```

```python
uriage_data["purchase_month"] = uriage_data["purchase_date"].dt.strftime("%Y%m")
uriage_data["purchase_month"].head()
```

```python
# aggfunc="size"で件数を集計
res = uriage_data.pivot_table(index="purchase_month", columns="item_name", aggfunc="size", fill_value=0)
res
```

```python
res = uriage_data.pivot_table(index="purchase_month", columns="item_name", values="item_price", aggfunc="sum", fill_value=0)
res
```

```python
# データの揺れを補正

# 小文字を大文字に変換
uriage_data["item_name"] = uriage_data["item_name"].str.upper()
print('length：',str(len(pd.unique(uriage_data["item_name"]))))
print(pd.unique(uriage_data["item_name"]))
# 半角の空文字を除去
uriage_data["item_name"] = uriage_data["item_name"].str.replace(" ", "")
print('length：',str(len(pd.unique(uriage_data["item_name"]))))
print(pd.unique(uriage_data["item_name"]))
# 全角の空文字を除去
uriage_data["item_name"] = uriage_data["item_name"].str.replace("　", "")
print('length：',str(len(pd.unique(uriage_data["item_name"]))))
print(pd.unique(uriage_data["item_name"]))
```

```python
# 欠損値の確認(axis=1は列で確認する)
uriage_data.isnull().any(axis=0)
```

```python
# item_priceで欠損値があるかを確認
flg_is_null = uriage_data["item_price"].isnull()
# 欠損値のある商品名をループする
for trg in list(uriage_data.loc[flg_is_null, "item_name"].unique()):
    # 欠損してないかつ欠損値のある商品名の金額を取得
    price = uriage_data.loc[(~flg_is_null) & (uriage_data["item_name"] == trg), "item_price"].max()
    # 欠損を起こしているデータに対して、保管を行う
    uriage_data["item_price"].loc[(flg_is_null) & (uriage_data["item_name"] == trg)] = price
uriage_data.head()
```

```python
# 全ての商品名に対してループする
for trg in list(uriage_data["item_name"].sort_values().unique()):
    # 商品毎の最大金額と最小金額が一致していることを確認
    print(trg + "の最大金額：" + str(uriage_data.loc[uriage_data["item_name"]==trg]["item_price"].max())+ ', 最小金額：' + str(uriage_data.loc[uriage_data["item_name"]==trg]["item_price"].min(skipna=False)))
```

```python
# 顧客名から全角と半角を削除
kokyaku_daicho_data["顧客名"] = kokyaku_daicho_data["顧客名"].str.replace(" ", "")
kokyaku_daicho_data["顧客名"] = kokyaku_daicho_data["顧客名"].str.replace("　", "")
```

```python
# 数値として取り込まれているものを日付に
flg_is_serial = kokyaku_daicho_data["登録日"].astype("str").str.isdigit()
from_Serial = pd.to_timedelta(kokyaku_daicho_data.loc[flg_is_serial, "登録日"].astype("float"), unit="D") + pd.to_datetime("1900/01/01")
from_Serial
```

```python
from_String = pd.to_datetime(kokyaku_daicho_data.loc[~flg_is_serial, "登録日"])
from_String
```

```python
# データの結合
kokyaku_daicho_data["登録日"] = pd.concat([from_Serial, from_String])
kokyaku_daicho_data
```

```python
# 登録年月別の顧客登録数
kokyaku_daicho_data["登録年月"] = kokyaku_daicho_data["登録日"].dt.strftime("%Y%m")
rslt = kokyaku_daicho_data.groupby("登録年月").count()["顧客名"]

print(rslt)
print(len(kokyaku_daicho_data))
```

```python
kokyaku_daicho_data.head()
```

```python
# データの結合（left_onとright_onでキーとなるデータを指定する）
join_data = pd.merge(uriage_data, kokyaku_daicho_data, left_on="customer_name", right_on="顧客名", how="left")
# customer_name列を削除
join_data = join_data.drop("customer_name", axis=1)
join_data
```

```python
# データをわかりやすく整形
dump_data = join_data[["purchase_date", "purchase_month", "item_name", "item_price",  "顧客名", "かな", "地域", "メールアドレス", "登録日"]]
# csvとして出力
dump_data.to_csv("dump_data.csv", index=False)
```
