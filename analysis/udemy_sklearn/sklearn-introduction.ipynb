```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
%matplotlib inline

iris = datasets.load_iris()
iris_data = iris.data
```

```python
st_data = iris_data[:50]
vc_data = iris_data[50:100]
vn_data = iris_data[100:150]

plt.scatter(st_data[:, 0], st_data[:, 1], label="Setosa")
plt.scatter(vc_data[:, 0], vc_data[:, 1], label="Versicolor")
plt.scatter(vn_data[:, 0], vn_data[:, 1], label="Versinica")
plt.legend()
plt.xlabel("Setal length")
plt.ylabel("Setal width")
plt.show()
plt.scatter(st_data[:, 2], st_data[:, 3], label="Setosa")
plt.scatter(vc_data[:, 2], vc_data[:, 3], label="Versicolor")
plt.scatter(vn_data[:, 2], vn_data[:, 3], label="Versinica")
plt.legend()
plt.xlabel("Petal length")
plt.ylabel("Petal width")
plt.show()
```

```python
from sklearn import datasets
from sklearn import svm

iris = datasets.load_iris()

# 分類機の作成
clf = svm.SVC()
# 訓練
clf.fit(iris.data, iris.target)

print(clf.predict([[5.1, 3.5, 1.4, 0.1], [6.5, 2.5, 4.4, 1.1], [5.9, 3.0, 6.2, 2.0]]))
```
