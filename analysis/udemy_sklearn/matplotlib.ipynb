```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,2,3,4,5])
y = np.array([3,5,9,2,8])
# グラフの描画
plt.plot(x,y)
plt.title('Line graph(MATLAB-style)')
plt.show()
```

```python
# 複数のグラフの描画

x = np.array([1,2,3,4,5])
y_1 = np.array([3,2,8,4,2])
y_2 = np.array([1,8,3,1,-1])

plt.plot(x, y_1, label="label1")
plt.plot(x, y_2, label="label2")
plt.xlabel('day')
plt.ylabel('item_name')
plt.legend(loc="best")
plt.show()
```

```python
# sin cosのカーブ
x = np.arange(0.0, 20.0, 0.1)
y_1 = np.sin(x)
y_2 = np.cos(x)

plt.plot(x, y_1, label="sin")
plt.plot(x, y_2, label="cos")
plt.legend(loc="best")
plt.show()
```

```python
x = np.array([1,2,3,4])
y = np.array([5,7,9,12])
labels = ['toru', 'marina', 'bob', 'tom']
# 棒グラフの描画（縦）
plt.bar(x, y, tick_label=labels)
plt.show()
```

```python
# 棒グラフの描画（横）
plt.barh(x, y, tick_label=labels)
plt.show()
```

```python
# ヒストグラム
x = np.random.normal(50, 10, 1000)
plt.hist(x)
plt.show()
```

```python
# 散布図
np.random.seed(1)
# ランダムな乱数を５０個生成
x = np.random.random(50)
y = np.random.random(50)
plt.scatter(x, y)
plt.show()
```

```python
# 散布図
np.random.seed(1)
# ランダムな乱数を５０個生成
x = np.random.random(50)
y = np.random.random(50)
# 下向き三角形
plt.scatter(x[0:10], y[0:10] ,marker='v', label='triangle down')
# 次の10個は上向き三角形
plt.scatter(x[10:20], y[10:20] ,marker='^', label='triangle up')
# 次の10個は上向き正方形
plt.scatter(x[20:30], y[20:30],marker='s', label='square')
# 次の10個は上向き正方形
plt.scatter(x[30:40], y[30:40] ,marker='*', label='star')
# 次の10個は上向き✖型
plt.scatter(x[40:50], y[40:50] ,marker='x', label='x')
plt.legend(loc="upper right")
plt.show()
```

```python
# 等高線
x = np.arange(3)
y = np.arange(5)

xx, yy = np.meshgrid(x, y)
plt.scatter(xx, yy)
plt.show()
```

```python
# データ作成
delta = 0.2
x = y = np.arange(-10.0, 10.0, delta)
xx, yy = np.meshgrid(x, y)
# z値の計算
z = xx ** 2 - yy**2
# 等高線の表示
plt.contour(xx, yy, z)
plt.show()
```

```python
# データ作成
delta = 0.2
x = y = np.arange(-10.0, 10.0, delta)
xx, yy = np.meshgrid(x, y)
# z値の計算
z = xx ** 2 - yy**2
# 等高線の表示
plt.contourf(xx, yy, z)
plt.show()
```

```python
# データの用意
x = np.array([1,2,3,4,5])
y_1 = np.array([11,12,23,12,56])
y_2 = np.array([15,2,13,12,16])
y_3 = np.array([11,12,23,12,56])
y_4 = np.array([8,5,22,13,8])

# 複数グラフを描画(横に並べる)
fig, axes = plt.subplots(ncols=2, nrows=2)
axes[0,0].plot(x, y_1)
axes[0,1].bar(x, y_2)
axes[1,0].barh(x, y_3)
axes[1,1].scatter(x, y_4)
plt.show()
```

```python
# 三次元のグラフの描画
delta = 0.2
x = y = np.arange(-10.0, 10.0, delta)
xx, yy = np.meshgrid(x,y)

# zの値を計算
z = xx**2 - yy**2

# 描画オブジェクトの生成(縦、横)
fig = plt.figure(figsize=(6,6))
# 三次元グラフ描画のサブプロット
ax = fig.gca(projection="3d")
# 等高線の表示
ax.contour3D(xx, yy, z, levels=40)

# x, y, zの範囲を設定
ax.set_xlim(xx.min()-1, xx.max()+1)
ax.set_ylim(yy.min()-1, yy.max()+1)
ax.set_zlim(-abs(z).max(), abs(z).max())

plt.savefig('contour3d.jpg', dpi=120)
plt.show()
```

```python
fig = plt.figure()
ax = fig.gca()

cset = ax.contour(xx, yy, z, levels=10)
# x, yの範囲を設定
ax.set_xlim(xx.min()-1, xx.max()+1)
ax.set_ylim(yy.min()-1, yy.max()+1)
# 等高線にラベルを付ける
ax.clabel(cset, fontsize=9, inline=1)

# 画像保存
plt.savefig('contour3d_2d.jpg', dpi=120)
plt.show()
```
